```python
from import_all import *
from utilities import *
import pyvisa
import trigger_fns

station = Station(config_file="electrochemistry.station.yaml")
keithley1 = station.load_keithley1()
keithley2 = station.load_keithley2()
```

```python
top_gate = { "channel": keithley2.smua,
        "name": "top_gate",
        "independent":  True, # default = True   
        "threshold_check": False,
        "threshold": 0,      
}

top_gate["voltage"] = np.linspace(-1,1,10)

chan = top_gate["channel"]
chan.nplc(5)
rm = pyvisa.ResourceManager()
intfc = rm.open_resource('GPIB0::INTFC')
```

```python
chan.write(f"{chan.channel}.trigger.source.linearv({volt}, 0.2, 10)")
chan.write(f"{chan.channel}.trigger.source.action = {chan.channel}.ENABLE")
chan.write(f"{chan.channel}.trigger.source.stimulus = trigger.EVENT_ID")
# chan.write(f"{chan.channel}.trigger.arm.stimulus = display.trigger.EVENT_ID")
chan.write(f"{chan.channel}.trigger.measure.i({chan.channel}.nvbuffer1)")
chan.write(f"{chan.channel}.nvbuffer1.clear()")
chan.write(f"{chan.channel}.nvbuffer1.appendmode = 1")
chan.write(f"{chan.channel}.measure.count = 1")
chan.write(f"{chan.channel}.trigger.measure.action = {chan.channel}.ENABLE")
chan.write(f"{chan.channel}.trigger.measure.stimulus = 0")
chan.write(f"{chan.channel}.trigger.count = 10")
chan.write(f"{chan.channel}.trigger.arm.count = 1")
chan.write(f"{chan.channel}.trigger.endsweep.action = smua.SOURCE_HOLD")

chan.write(f"{chan.channel}.trigger.initiate()")
```

```python
v = [0.1,0.2,0.3]
volt = 0.1
for volt in v:
    volt = str(volt)
    chan.write("smua.abort()")
    chan.write(f"{chan.channel}.trigger.source.linearv({volt}, {volt}, 1)")
    chan.write(f"{chan.channel}.trigger.source.action = {chan.channel}.ENABLE")
    chan.write(f"{chan.channel}.trigger.source.stimulus = trigger.EVENT_ID")

    chan.write(f"{chan.channel}.trigger.measure.i({chan.channel}.nvbuffer1)")
    chan.write("smua.nvbuffer1.clear()")
    chan.write("smua.nvbuffer1.appendmode = 1")
    chan.write(f"{chan.channel}.measure.count = 1")
    chan.write(f"{chan.channel}.trigger.measure.action = {chan.channel}.ENABLE")
    chan.write(f"{chan.channel}.trigger.measure.stimulus = 0")

    chan.write(f"{chan.channel}.trigger.count = 1")
    chan.write(f"{chan.channel}.trigger.arm.count = 1")
    

    chan.write(f"{chan.channel}.trigger.initiate()")

    sleep(1)
    intfc.assert_trigger()
    sleep(1)

    

```
